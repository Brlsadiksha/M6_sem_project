### Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from datetime import datetime


# model
from darts import TimeSeries
from darts.dataprocessing.transformers import Scaler
from darts.utils.timeseries_generation import datetime_attribute_timeseries
```

### data preprocessing_el

```python
# Load the electric dataset(CSV) into a Pandas DataFrame
df_el = pd.read_csv('el_price.csv', sep = ';')
# Select the desired columns
df_el = df_el[['MTU (CET/CEST)', 'Day-ahead Price [EUR/MWh]']]

# Rename the columns
df_el = df_el.rename(columns={'MTU (CET/CEST)': 'date', 'Day-ahead Price [EUR/MWh]': 'price'})

# Extract the first 16 characters from the 'date' column
df_el['date'] = df_el['date'].apply(lambda x: x[0:16])
    
# Interpolate missing values in the 'price' column using linear method
df_el['price'].interpolate(method='linear', inplace=True)

ts = TimeSeries.from_dataframe(df_el, time_col='date', value_cols = 'price', fill_missing_dates=True, freq=None)
```

### data preprocessing_weather

```python
# weather data import
df_weather = pd.read_csv('weather_2023.csv')

# define the input and output format
input_format = "%Y-%m-%dT%H:%M"
output_format = "%d.%m.%Y %H:%M"

# convert format
df_weather['time'] = df_weather['time'].apply(lambda x: datetime.strptime(x, input_format).strftime(output_format))

# nan data handling_interpolate(method='linear', inplace=True)
df_weather['temperature_2m (°C)'].interpolate(method='linear', inplace=True)
df_weather['direct_radiation (W/m²)'].interpolate(method='linear', inplace=True)
df_weather['diffuse_radiation (W/m²)'].interpolate(method='linear', inplace=True)
df_weather['windspeed_10m (m/s)'].interpolate(method='linear', inplace=True)
df_weather['windspeed_100m (m/s)'].interpolate(method='linear', inplace=True)

# rename
df_weather = df_weather.rename(columns={'time': 'date', 'windspeed_100m (m/s)' : 'wind', 'windspeed_10m (m/s)' : 'wind_10'})
df_weather = df_weather.rename(columns={'temperature_2m (°C)': 'temperature', 'direct_radiation (W/m²)' : 'radiation'})

# join the df_el and df_weather
df_total = pd.merge(df_el, df_weather, on ='date', how = 'inner')
```

### look into covariantes between price and weather variables.

```python
# change the data into arrary
price_array = df_total.price.values
wind_array = df_total.wind.values
temperature_array = df_total.temperature.values
radiation = df_total.radiation.values

# covariants
covariance_pw = np.cov(price_array, wind_array)[0, 1]
covariance_pt = np.cov(price_array, temperature_array)[0, 1]
covariance_pr = np.cov(price_array, radiation)[0, 1]
print("Covariance of price and wind :", covariance_pw)
print("Covariance of price and tempertures :", covariance_pt)
print("Covariance of price and radiation :", covariance_pr)
```

### covariants with date

```python
# create each series by date attribution

start_time = ts.start_time()  # Start time of the series
end_time = ts.end_time()  # End time of the series
freq_str = ts.freq_str  # Frequency string for the time series
num_periods = len(ts)  # Length of the time series

hour_series = datetime_attribute_timeseries(
    pd.date_range(start=ts.start_time(), freq=ts.freq_str, periods=len(ts)),
    attribute="hour",
    one_hot=False)
day_series = datetime_attribute_timeseries(
    pd.date_range(start=ts.start_time(), freq=ts.freq_str, periods=len(ts)),
    attribute="day",
    one_hot=False)
month_series = datetime_attribute_timeseries(
    pd.date_range(start=ts.start_time(), freq=ts.freq_str, periods=len(ts)),
    attribute="month",
    one_hot=False)
year_series = datetime_attribute_timeseries(
    pd.date_range(start=ts.start_time(), freq=ts.freq_str, periods=len(ts)),
    attribute="year",
    one_hot=False)
weekday_series = datetime_attribute_timeseries(
    pd.date_range(start=ts.start_time(), freq=ts.freq_str, periods=len(ts)),
    attribute="dayofweek",
    one_hot=False)
weekofyear_series = datetime_attribute_timeseries(
    pd.date_range(start=ts.start_time(), freq=ts.freq_str, periods=len(ts)),
    attribute="weekofyear",
    one_hot=False)

# series_transform
hour_trasform = Scaler().fit_transform(hour_series)
day_transform = Scaler().fit_transform(day_series)
month_transform = Scaler().fit_transform(month_series)
year_transform = Scaler().fit_transform(year_series)
weekday_transform = Scaler().fit_transform(weekday_series)
weekofyear_transform = Scaler().fit_transform(weekofyear_series)

# series stack
cov_all = hour_trasform.stack(day_transform).stack(month_transform).stack(year_transform).stack(weekday_transform).stack(weekofyear_transform)
```

```python
plt.figure(figsize=(15,5))
cov_all[:-240].plot()
```

```python
ts_price = TimeSeries.from_dataframe(df_total, time_col='date', value_cols = 'price', fill_missing_dates=True, freq=None)
ts_wind = TimeSeries.from_dataframe(df_total, time_col='date', value_cols = 'wind', fill_missing_dates=True, freq=None)
```

```python
from darts import concatenate
from darts.utils.timeseries_generation import datetime_attribute_timeseries as dt_attr

price_covs = concatenate(
    [
        dt_attr(ts_price.time_index, "month", dtype=np.float32) / 5,
        (dt_attr(ts_price.time_index, "year", dtype=np.float32) - 2023),
    ],
    axis="component",
)

wind_covs = concatenate(
    [
        dt_attr(ts_wind.time_index, "month", dtype=np.float32) / 5,
        (dt_attr(ts_wind.time_index, "year", dtype=np.float32) - 2000),
    ],
    axis="component",
)

price_covs.plot()
plt.title(
    "one multivariate time series of 2 dimensions, containing covariates for the ts_price:"
);
```
